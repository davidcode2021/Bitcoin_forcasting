```python
import pandas as pd
```

```python
df=pd.read_csv("Bitcoin.csv",sep=";")
df
```

```python
df1=df.loc[:,"timeLow":"volume"]
df1
```

```python
def get_date(date):
    date=date.split("T")
    return date[0]

df1["timeLow"]=df1["timeLow"].apply(get_date)
df1.rename(columns={"timeLow":"date"},inplace=True)
```

```python
df1
```

```python
for x in df1["date"].duplicated():
    if x=='True':
        print('yes')
    else:
        print("No")
```

```python
df1.loc[df1.index==286,"date"]="2022-12-14"
df1.iloc[270:290,:]
```

```python
df1.index=pd.to_datetime(df1["date"])
df1.drop(columns="date",inplace=True)
```

```python
df1
```

```python
df1['close'].plot()
```

```python
df1["tomorrow"]=df1["close"].shift(-1)
```

```python
df1.dropna(inplace=True)
```

```python
df1['PE_inc']=(df1["tomorrow"]-df1['close'])/df1["close"] * 100
```

```python
X=df1.iloc[:,:5]
Y=df1.iloc[:,6]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(X)
X
```

```python
features=["open","high","low","close","volume"]
index=pd.date_range(start="2022-12-03",end="2023-09-26")

X=pd.DataFrame(X,columns=features,index=index)
X.sort_index(axis=0,ascending=False,inplace=True)
```

```python
data=pd.concat([X,df1["PE_inc"]],axis=1)
# df1.index
data
```
